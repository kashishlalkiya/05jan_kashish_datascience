```python
#10. Create an array of numbers between 0 and 6 with increments of 0.3 and name its “x”. Then on the same plot, plot x, x², x³, and x⁴. For consistency, use the following style lines respectively, “ro”, “bs”, “g”, and “:”. Lastly, make sure that the x-axis covers 0 to 6, whilethe y-axis spans from 0 to 125. Do not worry if you are familiar with the style lines — you will recognize them as soon as you see the plot.
import numpy as np
x = np.arange(0, 6.3, 0.3)
plt.plot(x, x, 'ro', label='x')
plt.plot(x, x**2, 'bs', label='x²')
plt.plot(x, x**3, 'g', label='x³')
plt.plot(x, x**4, ':', label='x⁴')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([0, 6, 0, 125])
plt.legend()
plt.show()
```

```python
# 11. Heights and initials of a group of individuals are provided below. Create a bar plot titled “Height Comparison” to compare the heights among this group.
#height = [179, 155, 191, 152, 188, 177]
#names = ['QA', 'WB', 'EC', 'RD', 'TE', 'YF'] 
heights = [179, 155, 191, 152, 188, 177]
names = ['QA', 'WB', 'EC', 'RD', 'TE', 'YF']
plt.bar(names, heights, color='skyblue')
plt.xlabel('Names')
plt.ylabel('Height (cm)')
plt.title('Height Comparison')
plt.show()
```

```python
#12. Plot a histogram of x, where x consists of 100,000 randomly-selected points with a normal distribution (hint: you can use numpy.random.randn() to generate the random points). The histogram should have 10 bins. Look at how the histogram changes when we try 20 and 50 bins.
x = np.random.randn(100000)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.hist(x, bins=10, color='skyblue', edgecolor='black')
plt.title('10 Bins')
plt.subplot(1, 3, 2)
plt.hist(x, bins=20, color='skyblue', edgecolor='black')
plt.title('20 Bins')
plt.subplot(1, 3, 3)
plt.hist(x, bins=50, color='skyblue', edgecolor='black')
plt.title('50 Bins')
plt.tight_layout()
plt.show()
```
